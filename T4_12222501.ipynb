```python
import numpy as np
class Game:
    def __init__(self):
        self.a_easy=np.random.randint(100)
        self.a_hard=np.random.randint(1000)
        print("1. Easy")
        print("2. Hard")
        print("3. Exist")
        self.choice=int(input("Enter choice "))
        self.i=0
        if self.choice==1:
            self.l=6
            self.start()
        elif self.choice==2:
            self.l=9
            self.start()
        elif self.choice==3:
            print("Thank you")
            return
        else:
            print("Invalid")  
    def start(self):
        self.accept()
    def accept(self):
            g=int(input("Enter you guess: "))
            self.g=g
            self.validate(g)
    def validate(self,g):
        if self.choice==1:
            if g==self.a_easy:
                self.flag=1
            else:
                self.flag=0
        if self.choice==2:
            if g==self.a_hard:
                self.flag=1
            else:
                self.flag=0
        self.hint()
    def hint(self):
        self.i+=self.i
        if self.flag==1:
            print("Correct")
            if self.choice==1:
                print(self.a_easy)
            if self.choice==2:
                print(self.a_hard)
    
        else:
            if self.choice==1:
                if self.g>self.a_easy:
                    print("Smaller")
                else:
                    print("Greater")
            if self.choice==2:
                if self.g>self.a_hard:
                    print("Smaller")
                else:
                    print("Greater")
    
            if self.i<self.l:
                self.start()
g=Game()
```
