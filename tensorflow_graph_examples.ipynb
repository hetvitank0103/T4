## Quadratic Function using TensorFlow Graph
Compute f(x) = axÂ² + bx + c using `@tf.function`. This defines a TensorFlow graph.

```
import tensorflow as tf

# Define the graph function for quadratic equation
@tf.function
def quadratic(a, b, c, x):
    return a * x**2 + b * x + c

# Define constants
a = tf.constant(2.0)
b = tf.constant(3.0)
c = tf.constant(1.0)
x = tf.constant(4.0)

# Execute the graph
y = quadratic(a, b, c, x)

print("Quadratic output f(x):", y.numpy())  # Expected: 2*4^2 + 3*4 + 1 = 45.0
```

## Area and Perimeter of a Rectangle using TensorFlow Graph

```
import tensorflow as tf

# Define the graph function for area and perimeter
@tf.function
def compute_area_perimeter(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter

# Define constants
length = tf.constant(5.0)
width = tf.constant(3.0)

# Execute the graph
area, perimeter = compute_area_perimeter(length, width)

print("Length:", length.numpy())
print("Width:", width.numpy())
print("Area:", area.numpy())
print("Perimeter:", perimeter.numpy())
```
