```python
# Single
class parent:
    def fun(self):
        print("Parent Class")
        
class child(parent):
    def Cfun(self):
        print("Child Class")
        
ob = child()
ob.Cfun()
ob.fun()
```

```python
# MultiLevel
class Gparent:
    def fun(self):
        print("Grand Parent Class")

class parent(Gparent):
    def fun(self):
        super().fun()
        print("Parent Class")
        
class child(parent):
    def Cfun(self):
        print("Child Class")
        
ob = child()
ob.Cfun()
ob.fun()
```

```python
# Multiple

class parent1:
    def fun1(self):
        print("Parent1 Class")

class parent2:
    def fun2(self):        
        print("Parent2 Class")
        
class child(parent1,parent2):
    def Cfun(self):
        print("Child Class")
        
ob = child()
ob.Cfun()
ob.fun1()
ob.fun2()
```

```python
# Hierarchical
class parent1:
    def fun1(self):
        print("Parent1 Class")

class child1(parent1):
    def Cfun1(self):        
        print("Child1 Class")
        
class child2(parent1):
    def Cfun2(self):
        print("Child2 Class")
        
ob = child1()
ob.Cfun1()
#ob.Cfun2()
ob.fun1()
```

```python
# Multiple

class parent1:
    def fun(self):
        print("Parent1 Class")

class parent2:
    def fun(self):        
        print("Parent2 Class")
        
class child(parent2,parent1):
    def Cfun(self):
        print("Child Class")
        
ob = child()
ob.Cfun()
ob.fun()
ob.fun()

ob.fun()

print(child.__mro__)
```

```python
#Operator Overloading

class Point:
    def __init__(self,x):
        self.x=x
        
    def __sub__(self,other):
        return self.x - other.x
    
    def __add__(self,other):
        return self.x + other.x
    
    def __lt__(self,o):
        if self.x<o.x:
            print(self.x,"is less than ",o.x)
        else:
            print(o.x,"is less than ",self.x)
o1 = Point(5)
o2 = Point(10)
print("Addition: ",o1+o2)
print("Sub: ",o1-o2)
o1<o2
```

```python
#653
from abc import ABC,abstractmethod
import math

@Bclass Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    
class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
        
    def calculate_area(self):
        self.area_of_rectangle = (self.l * self.b)
        print("Area of rectangle: ",self.area_of_rectangle)
        
    def __gt__(self,o):
        if self.area_of_rectangle >o.area_of_rectangle:
            print("Rect1 has greater area than Rect2")
        else:
            print("Rect2 has greater area than Rect1")
        
class Circle(Shape):
    def __init__(self,r):
        self.r=r
        
    def calculate_area(self):
        self.area_of_circle =(math.pi*self.r)**2
        print("Area of circle: ",self.area_of_circle)
 
rect1 = Rectangle(10,20)
rect2 = Rectangle(5,2)
rect1.calculate_area()
rect2.calculate_area()
rect1>rect2
print("---------------------------------------------")
c1 = Circle(20)
c1.calculate_area()
print(Circle.__mro__)
```

```python
#651
from abc import ABC, abstractmethod
class Employee(ABC):
    @abstractmethod
    def receive_call(self):
        pass
  
    @abstractmethod
    def end_call(self):
        pass
    @abstractmethod
    def is_free(self):
        pass
  
    @abstractmethod
    def get_rank(self):
        pass
    
class Respondent(Employee):
    def __init__(self,id,name):
        self.id=id
        self.name=name
        self.rank=3
        self.free=True
        
    def receive_call(self):
        print("Call received by",self.name)
        self.free=False
        
    def end_call(self):
        print("Call ended")
        self.free=True
        
    def is_free(self):
        return self.free
    
    def get_rank(self):
        return self.rank
    
class Manager(Employee):
    def __init__(self,id,name):
        self.id=id
        self.name=name
        self.rank=2
        self.free=True
        
    def receive_call(self):
        print("Call received by",self.name)
        self.free=False
        
    def end_call(self):
        print("Call ended")
        self.free=True
        
    def is_free(self):
        return self.free
    
    def get_rank(self):
        return self.rank
    
class Director(Employee):
    def __init__(self,id,name):
        self.id=id
        self.name=name
        self.rank=1
        self.free=True
        
    def receive_call(self):
        print("Call received by",self.name)
        self.free=False
        
    def end_call(self):
        print("Call ended")
        self.free=True
        
    def is_free(self):
        return self.free
    
    def get_rank(self):
        return self.rank
class Call:
    def __init__(self,Id,name):
        self.Id=Id
        self.name=name
        self.assigned = False
    
class CallHandler:
    res=[]
    manager=[]
    director=[]
    
    def add_employee(self,ob):
        if ob.rank==1:
            CallHandler.director.append(ob)
        elif ob.rank==2:
            CallHandler.manager.append(ob)
        elif ob.rank==3:
            CallHandler.res.append(ob)
            
    def dispatch_call(self,call):
        for i in CallHandler.res:
            if i.is_free():
                i.receive_call()
                call.assigned = True
                break
        else:
            for i in CallHandler.manager:
                if i.is_free():
                    i.receive_call()
                    call.assigned = True
                    break
            else:
                for i in CallHandler.director:
                    if i.is_free():
                        i.receive_call()
                        call.assigned = True
                        break
                else:
                    print("Sorry! All employees are currently busy.")
                    
R1 = Respondent(1,"R1")
R2 = Respondent(2,"R2")
R3 = Respondent(3,"R3")

M1 = Manager(11,"M1")
M2 = Manager(12,"M2")

D1 = Director(100,"D1")

ch = CallHandler()
ch.add_employee(R1)
ch.add_employee(R2)
ch.add_employee(R3)
ch.add_employee(M1)
ch.add_employee(M2)
ch.add_employee(D1)

C1 = Call(101,"Ram")
C2 = Call(102,"Rama")
C3 = Call(103,"Aham")
C4 = Call(104,"Alice")

ch.dispatch_call(C1)
ch.dispatch_call(C2)
ch.dispatch_call(C3)
ch.dispatch_call(C4)
```
