This notebook is designed for classroom teaching and includes practice examples with solutions.


```
import tensorflow as tf

# Create tensors with different types and shapes
scalar = tf.constant(42)
vector = tf.constant([1.0, 2.0, 3.0])
matrix = tf.constant([[1, 2], [3, 4]])
tensor_3d = tf.constant([[[1], [2]], [[3], [4]]])

print("Scalar:", scalar.numpy())
print("Vector:", vector.numpy())
print("Matrix:\n", matrix.numpy())
print("3D Tensor:\n", tensor_3d.numpy())

# Data types
print("Data type of matrix:", matrix.dtype)
print("Shape of tensor_3d:", tensor_3d.shape)
```

```
# Slice rows and elements
tensor = tf.constant([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Original Tensor:\n", tensor.numpy())

# First row
print("First row:", tensor[0].numpy())

# Last column
print("Last column:", tensor[:, -1].numpy())

# Middle 2x2 submatrix
print("Middle 2x2:\n", tensor[0:2, 1:3].numpy())
```

```
# Reshape a 1D tensor into 2D
t = tf.range(12)
reshaped = tf.reshape(t, (3, 4))
print("Original 1D tensor:", t.numpy())
print("Reshaped to 3x4:\n", reshaped.numpy())
```

```
# Broadcasting addition
a = tf.constant([[1], [2], [3]])
b = tf.constant([10, 20, 30])
print("Tensor A:\n", a.numpy())
print("Tensor B:", b.numpy())
result = a + b
print("Broadcasted addition result:\n", result.numpy())
```

```
# Type conversion (float to int)
float_tensor = tf.constant([1.5, 2.8, 3.3], dtype=tf.float32)
int_tensor = tf.cast(float_tensor, tf.int32)
print("Original float tensor:", float_tensor.numpy())
print("Converted to int:", int_tensor.numpy())
```
