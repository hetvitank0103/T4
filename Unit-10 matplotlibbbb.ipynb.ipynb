```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[5,2,8,7]
plt.title("My graph",fontsize=15,color="red",fontweight="bold")
plt.xlabel("x axis",fontsize=12,color="green",fontweight="bold")
plt.ylabel("y axis",fontsize=12,color="green",fontweight="bold")
plt.plot(x,y,color="blue",linestyle=":",linewidth=4,label="x vs y") # line styles=:,--,-.
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)
y=np.arange(5,15)
z=np.arange(10,30,2)
plt.title("My graph",fontsize=15,color="red",fontweight="bold")
plt.xlabel("x axis",fontsize=12,color="green",fontweight="bold")
plt.ylabel("y axis",fontsize=12,color="green",fontweight="bold")
plt.plot(x,y,color="blue",linestyle=":",linewidth=4,label="x vs y",marker="*",ms=10) # line styles=:,--,-.
plt.plot(x,z,color="orange",linestyle="-.",linewidth=4,label="x vs z",marker="o",ms=10)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)
y=np.arange(5,15)
z=np.arange(10,30,2)
plt.title("My graph",fontsize=15,color="red",fontweight="bold")
plt.xlabel("x axis",fontsize=12,color="green",fontweight="bold")
plt.ylabel("y axis",fontsize=12,color="green",fontweight="bold")
plt.plot(x,y,"o:m",label="x vs y") # line styles=:,--,-.
plt.plot(x,z,"^:y",label="x vs z") # m,r,c,w,y,k-black,b,g
plt.legend()
plt.show()
```

# sub plots
 

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,11)
y=np.arange(5,15)
z=np.arange(10,30,2)

plt.subplot(1,2,1)
plt.plot(x,y,"d:r")

plt.subplot(1,2,2)
plt.plot(x,z,"*-.g")

plt.suptitle("subplots")
plt.show()
```

# bar plot

```python
student=["a","b","c","d"]
marks=[10,20,24,16]
plt.bar(student,marks,color=["red","green","blue","black"],width=0.14)
plt.show()
```

```python
student=["a","b","c","d"]
marks=[10,20,24,16]
plt.title("bar graph")
plt.xlabel("student")
plt.ylabel("marks")
plt.bar(student,marks,color=["red","green","blue","black"],width=0.14,label="student vs marks")
plt.legend()
plt.show()
```

```python
student=["a","b","c","d"]
marks=[10,20,24,16]
plt.title("bar graph")
plt.ylabel("student")
plt.xlabel("marks")
plt.barh(student,marks,color=["red","green","blue","black"],label="student vs marks")
plt.legend()
plt.show()
```

# pie chart

# area plot

```python
x=[1,2,3,4,5]
y=[1,4,8,5,2]
plt.plot(x,y,color="red")
plt.fill_between(x,y,alpha=0.4)
plt.show()
```

```python
# stack plot
x=[1,2,3,4,5]
y1=[1,4,8,5,2]
y2=[11,14,18,15,12]
y3=[31,24,18,35,22]
plt.stackplot(x,y1,y2,y3,labels=["a","b","c"],colors=["red","green","blue"],
              alpha=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
month=np.arange(1,13)
income=np.array([40000,45000,48000,43000,44000,51000,40000,45000,48000,
                 43000,44000,51000])
expense=np.array([43000,44000,51000,40000,45000,48000,43000,44000,51000,
                  40000,45000,48000])
plt.plot(month,income,color="green")
plt.plot(month,expense,color="red")
plt.fill_between(month,income,expense,where=(income>expense),color="green",
                 label="positive",interpolate=True)
plt.fill_between(month,income,expense,where=(income<=expense),color="red",
                 label="negative",interpolate=True)
plt.xlabel("month")
plt.ylabel("income/expense")
plt.legend()
plt.show()
```
