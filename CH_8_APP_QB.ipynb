 Implement a python program that performs basic arithmetic operations using TensorFlow.

```python
import tensorflow as tf

# Define two constant tensors
a = tf.constant(10)
b = tf.constant(5)

# Perform arithmetic operations
add_result = tf.add(a, b)
sub_result = tf.subtract(a, b)
mul_result = tf.multiply(a, b)
div_result = tf.divide(a, b)

# Print results
print("Tensor a:", a.numpy())
print("Tensor b:", b.numpy())
print("Addition (a + b):", add_result.numpy())
print("Subtraction (a - b):", sub_result.numpy())
print("Multiplication (a * b):", mul_result.numpy())
print("Division (a / b):", div_result.numpy())
```

 Implement a python program that performs matrix operations using TensorFlow.

```python
import tensorflow as tf

# Define two 2x2 matrices
A = tf.constant([[1.0, 2.0],
                 [3.0, 4.0]])

B = tf.constant([[5.0, 6.0],
                 [7.0, 8.0]])

# Matrix addition
add_result = tf.add(A, B)

# Matrix subtraction
sub_result = tf.subtract(A, B)

# Matrix multiplication (dot product)
matmul_result = tf.matmul(A, B)

# Transpose of A
transpose_A = tf.transpose(A)

# Determinant of A
det_A = tf.linalg.det(A)

# Inverse of A (only if determinant â‰  0)
inverse_A = tf.linalg.inv(A)

# Print results
print("Matrix A:\n", A.numpy())
print("Matrix B:\n", B.numpy())
print("Addition (A + B):\n", add_result.numpy())
print("Subtraction (A - B):\n", sub_result.numpy())
print("Multiplication (A * B):\n", matmul_result.numpy())
print("Transpose of A:\n", transpose_A.numpy())
print("Determinant of A:", det_A.numpy())
print("Inverse of A:\n", inverse_A.numpy())
```

Write python to find the number of zeros at the end of a factorial of a given positive number using tensorflow.

```python
import tensorflow as tf

@tf.function
def trailing_zeros_in_factorial(n):
    count = tf.constant(0, dtype=tf.int32)
    i = tf.constant(5, dtype=tf.int32)

    # Loop while n // i > 0
    while tf.math.greater(n // i, 0):
        count += n // i
        i *= 5

    return count

# Example usage
n = tf.constant(75, dtype=tf.int32)
result = trailing_zeros_in_factorial(n)
tf.print("Number of trailing zeros in", n, "factorial is:", result)
```

Using tensorflow Write a Python function to check whether a number is perfect or not.
 The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1
 + 2 + 3 = 6. Equivalently, the number 6 is equal to half the sum of all its positive
 divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6. The next perfect number is 28 = 1 + 2 + 4 + 7 + 14.
 This is followed by the perfect numbers 496 and 8128.

```python
import tensorflow as tf

@tf.function
def is_perfect_number(n):
    total = tf.constant(0, dtype=tf.int32)

    # Loop from 1 to n//2 to find proper divisors
    for i in tf.range(1, n // 2 + 1):
        if n % i == 0:
            total += i

    # Return True if the sum of divisors equals n
    return tf.equal(total, n)

# Test the function
test_numbers = [6, 28, 496, 8128, 10, 12, 97]

for num in test_numbers:
    result = is_perfect_number(tf.constant(num))
    tf.print("Is", num, "a perfect number?", result)
```

Using tensorflow Write a python program to check the validity of password without using any built-in
 functions or modules. Password checker program basically checks if a password is
 valid or not based on the password policies mention below:
 â€¢ Password should contain at least one lowercase letter(a-z).
 â€¢ Password should contain at least one uppercase letter(A-Z).
 â€¢ Password should contain at least one special character (@, #, %, &amp;, !, $, etcâ€¦).
 â€¢ Password should not contain any space.
 â€¢ Password should contain at least one digit (0-9).
 â€¢ Password length should be between 8 to 15 characters.
 â€¢ It should not contain the repeated combination of consecutive 3 characters.
 Also check whether the password is strong or not without using any built-in function.
 Given a string, find its strength. Let a strong password is one that satisfies all above
 conditions. A moderate password is one that satisfies first three conditions and has
 length at least 6. Otherwise password is weak.

```python
import tensorflow as tf

def check_password_easy(password):
    chars = tf.constant([ord(c) for c in password], dtype=tf.int32)
    length = tf.shape(chars)[0]

    has_lower = has_upper = has_digit = has_special = has_space = False

    specials = [33, 35, 36, 37, 38, 64]  # ! # $ % & @

    for c in chars.numpy():
        if 97 <= c <= 122:  # a-z
            has_lower = True
        elif 65 <= c <= 90:  # A-Z
            has_upper = True
        elif 48 <= c <= 57:  # 0-9
            has_digit = True
        elif c in specials:
            has_special = True
        elif c == 32:  # space
            has_space = True

    # Rule: no repeated 3-character combination
    no_repeat = True
    s = password
    for i in range(len(s) - 5):
        if s[i:i+3] == s[i+3:i+6]:
            no_repeat = False
            break

    # All conditions
    valid = (8 <= len(password) <= 15 and
             has_lower and has_upper and has_digit and
             has_special and not has_space and no_repeat)

    # Strength evaluation
    if valid:
        strength = "Strong"
    elif has_lower and has_upper and has_special and len(password) >= 6:
        strength = "Moderate"
    else:
        strength = "Weak"

    return valid, strength

# âœ… Test Examples
passwords = ["Pass@123", "weak1", "NoSpecial123", "Perfect@123!", "aaaAAA111!!!", "abcABC123abcABC123"]

for pwd in passwords:
    valid, strength = check_password_easy(pwd)
    print(f"Password: {pwd:20} | Valid: {valid} | Strength: {strength}")
```

Using tensorflow write a python code to Take that list and finds all pairs of integers that differ by three.

```python
import tensorflow as tf

def find_symmetric_pairs_diff_by_three(input_list):
    nums = tf.constant(input_list, dtype=tf.int32)
    pairs = []

    for i in range(len(input_list)):
        for j in range(len(input_list)):
            if i != j:
                diff = tf.abs(nums[i] - nums[j])
                if diff == 3:
                    pairs.append((nums[i].numpy(), nums[j].numpy()))

    return pairs

# ðŸ”Ž Test
input_list = [1, 4, 7, 10, 2, 5, 8]
pairs = find_symmetric_pairs_diff_by_three(input_list)

print("Fully symmetric pairs with difference of 3:")
for p in pairs:
    print(p)
```
